```python
import pandas as pd
import glob
import os

# Path to folder containing CSV files
csv_folder = '/Users/ekowaddai/Documents/GitHub/Geojson/GEO CSVS'  # e.g., './data'

# Use glob to get all CSV files in the folder
csv_files = glob.glob(os.path.join(csv_folder, '*.csv'))

# Load and combine all CSVs into one DataFrame
df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)

df.rename(columns={
    'shapeName': 'District'
}, inplace=True)

# Display the result
df.head(2000)
```

```python
count = df[df['District'].str.contains('district', case=False, na=False)].shape[0]
print("Rows containing 'District':", count)
```

```python
# Load your second CSV
df2 = pd.read_csv('/Users/ekowaddai/Documents/GitHub/Geojson/P1 District Level Geoboundaries - test.csv')

# Check it loaded
df2.head(1200)
```

```python
df['District'] = df['District'].str.replace(r'\s+District$', '', regex=True).str.strip()
count = df[df['District'].str.contains('district', case=False, na=False)].shape[0]
print("Rows containing 'District':", count)
```

```python
# Merge df with df2 on 'District' column
df_merged = df.merge(df2[['District', 'Region', 'Country']], on='District', how='left')

# Check result
df_merged.head(1400)
```

```python
# Rows where Region is still missing (i.e., no match found)
unmatched = df_merged[df_merged['Region'].isna()]

# Display them
print(unmatched[['District']])
```
